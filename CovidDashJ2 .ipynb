```python
import plotly.graph_objects as go
import plotly.offline as pyo
import dash
from dash import html
import requests
import pandas as pd
from dash import dcc
```

```python
#Fetch data from API
api_url = 'https://data.cdc.gov/resource/9mfq-cb36.json'
response = requests.get(api_url)
data = response.json()

# Prepare the data for Visualization
df = pd.DataFrame(data)  # Convert the data to a DataFrame
```

```python

# Create the dropdown options for States
states = df['state'].unique()
dropdown_options = [{'label': state + "- USA", 'value': state} for state in states]
```

```python
# Fetch dates from data and Create date filter
dates = df['submission_date'].unique()
dropdown_date_options = [{'label': date, 'value': date} for date in dates]
```

```python
#Define you app layout 
app = dash.Dash()

# Defining the HTML components

app.layout = html.Div(
    children=[
        html.H1("Covid-19 Dashboard", style={'color': 'red', 'text-align': 'center'}),
        html.Div(
            [dcc.Dropdown(id='state-dropdown', options=dropdown_options, value=states[0]),
            dcc.Dropdown(id='date-dropdown', options=dropdown_date_options, value=dates[0])],
            style={'color': 'red', 'text-align': 'left', 'display': 'inline-block', 'width': '40%'}
        ),
        html.Div(
            children=[
                dcc.Graph(id='cases-graph')
            ]
        )
    ]
)



# Create the callback function to update the chart
@app.callback(
    dash.dependencies.Output('cases-graph', 'figure'),
    [dash.dependencies.Input('state-dropdown', 'value'),
     dash.dependencies.Input('date-dropdown', 'value')]
)


def update_chart(state, date):
    # Filter the data based on the selected state and date
    filtered_data = df[(df['state'] == state) & (df['submission_date'] == date)]
    
    # Create the chart using Plotly Express or Plotly graph objects
    # Replace the following code with your desired chart creation logic
    fig = {
        'data': [
            {'x': filtered_data['submission_date'], 'y': filtered_data['new_case'], 'type': 'bar', 'name': 'New Cases'},
            {'x': filtered_data['submission_date'], 'y': filtered_data['tot_cases'], 'type': 'bar', 'name': 'Total Cases'},
            {'x': filtered_data['submission_date'], 'y': filtered_data['tot_death'], 'type': 'bar', 'name': 'Total Deaths'},
        ],
        'layout': {
            'title': f'Covid-19 Cases & Deaths in {state} - USA'
        }
    }
    
    return fig





if __name__ == '__main__':
    app.run(debug=True)
```
